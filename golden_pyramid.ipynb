
Consider a triangular pyramid $ABCD$ with the base $ABC$ and height 1.
Suppose that the points $A, B$ and $C$ are contained in the plane $\{z=0\}$.
Given the $xy$-coordinates of the points $A, B$ and $C$
find the $xy$-coordinates of $D$ such that the surface area of $ABCD$
is minimal.

<em>Input data.</em> The array <code>v</code> with shape (3,2) 
containes the $xy$-coordinates <code>v[0], v[1], v[2]</code>
of the vertices $A,B,C$ respectively.

<em>Output data.</em> The array <code>x</code> should contain the $xy$-cooridnates of the point $D$.


```python
%load_ext stepikutils.submitter
%load_dataset datasets/golden_pyramid.npz
%whos 
```

```python
%%solution_code 3219537

import numpy as np
x = np.zeros(2)
```

```python
from matplotlib.pyplot import *
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
```

```python

scatter(v[:,0], v[:,1])
scatter(x[0], x[1])
names = 'ABC'
for i in range(3):
    offset = np.array([0.05, 0.05])
    annotate(names[i], v[i] + offset, fontsize=14)
ax = gca()
ax.set_aspect('equal')
triangle = Polygon(v, color='b', fill=False)
ax.add_patch(triangle)
ax.set_axis_off()
# savefig('golden_pyramid.svg', format='svg')
show()
```
