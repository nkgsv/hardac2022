<p>
Given a signal of the form
$$
    x(t) = A \exp(i(\omega t + \varphi)) + \varepsilon(t),
$$
where $\varepsilon=\varepsilon(t)$ is some random noise, estimate the constants $A$, $\omega$ and $\varphi$.
</p>

<p><em>Input data.</em>
The variables <code>t</code> and <code>x</code> of type <code>numpy.ndarray</code> contain respectively
the values $t_k=\frac{k}{N}$ and $x(t_k)=\frac{k}{N}$, where $k=0,1,2,\ldots,N-1$.
</p>

<p><em>Output data.</em>
The estimated values of $A$, $\omega$ and $\varphi$ should be assigned respectively
to the variables <code>A</code>, <code>omega</code> and <code>phi</code> (of type <code>float</code>).
</p>
