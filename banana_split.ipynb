
<p>
Consider a signal of the form $x(t) = A \cdot \exp(i\omega(t) t) + \varepsilon(t)$,
where
$$
\omega(t) = \begin{cases}
\omega_1, & t \le t_c, \\
\omega_2, & t > t_c.
\end{cases}
$$
Here $A \in \mathbb C$ is a nonzero constant, $\omega_1$ and $\omega_2$ are some distinct nonzero real numbers,
$t_c > 0$ is constant and $\varepsilon=\varepsilon(t)$ is some random noise.
</p>

<p>
Given a sample $\{x(t_k)\}_{k=0}^{N-1}$ of the signal $x$ at times $t_k = \frac{k}{N} T$ ($T>0$), estimate the parameter $t_c$.
</p>

<p>
<em>Input data.</em> The array <code>x</code> contains the values $\{x(t_k)\}_{k=0}^{N-1}$ and the array <code>t</code> contains the values $\{t_k\}_{k=0}^{N-1}$.
The variable <code>T</code> contains the value of $T$.
</p>
<p>
<em>Output data.</em> The estimated value of $t_c$ should be assigned to the variable <code>t_c</code>.
</p>
